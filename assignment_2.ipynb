## Insights / observations

1. Bar chart represents more male was boarded in titanic as compared with female.
2. Men ranges between around 550 to 580.

```python
plt.pie(freq.values,labels=freq.index, autopct='%.2f%%')
plt.title('Percentage analysis Male vs Female ')
plt.show()
```

## Insights / observations 
above same data analysed through pie chart
1. Totally titanic was crowded with male as compared with female.
2. Almost 64.76% of the passagers is male and only 35.24% are females.

```python
sns.barplot(data=data,x='sex',y='survived')
plt.xlabel("Sex")
plt.ylabel("Survived")
plt.title("Count of survived people based on their sex")
plt.show()
```

## Insights / observations

1. Shows the count of people who survived based on their sex.
2. The female has survived more than male.

```python
sns.barplot(data=data,x='pclass',y='survived')
plt.xlabel("Passenger class")
plt.ylabel("Survived")
plt.title("Count of survived people based on their passenger class")
plt.show()
```

## Insights / observations

1. 1st class passengers are more safer as the number of survived is greater than the other classed.
2. 3rd class passengers are less secure as compared with other classes.

```python
sns.scatterplot(data=data,x='age',y='survived')
plt.xlabel("Age")
plt.ylabel("Survived")
plt.title("Count of survived people based on passenger age")
plt.show()
```

## Insights / observations
1. Here depicts that age is not a factor, as it affect equally.

```python
sns.barplot(data=data,x='sibsp',y='survived')
plt.xlabel("sibsp")
plt.ylabel("Survived")
plt.title("Count of survived people based on their sibsp")
plt.show()
```

## Insights / observations
1. The single sibling / spouse have higher number of survival rate.

```python
sns.barplot(data=data,x='parch',y='survived')
plt.xlabel("Parch")
plt.ylabel("Survived")
plt.title("Count of survived people based on their parch")
plt.show()
```

## Insight / observation
1. parch(Number of Parents/Children Aboard) number 3 has survived more.
2. parch(Number of Parents/Children Aboard) number 5 has survived less.

```python
sns.violinplot(x ="sex", y ="age", hue ="survived", data = data, split = True)
plt.title("Survived analysis based on their age and sex")
plt.show()
```

## Insight / observation
1. Survival rate is high for women between the age group 20 to 50.
2. Men survival rate decreases when age is increasing.

```python
sns.catplot(x ='embarked', hue ='survived', kind ='count', col ='pclass', data = data)

plt.show()
```

## Insight / observation
1. The majority of people boarded from Southampton (S).
2. The majority of class 3 people boarded from Queenstown (Q).
